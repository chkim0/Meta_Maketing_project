```python
df = pd.read_csv('csv/test.csv')
```

```python
import re

def remove_special_characters(text):
    # 이모티콘 및 특수문자 제거를 위한 정규표현식
    pattern = r'[^a-zA-Z0-9가-힣\s]'  # 영문, 한글, 숫자, 공백만 허용

    # 정규표현식을 사용하여 특수문자 제거
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text

# '내용' 열의 각 행에 대해 이모티콘과 특수문자 제거
df['contents'] = df['내용'].dropna().apply(remove_special_characters)
```

# TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 문장 리스트
sentences = df['contents'].dropna().tolist()

# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

# 특정 키워드의 TF-IDF 값을 확인
keyword_index = vectorizer.vocabulary_['빅데이터']
tfidf_values = tfidf_matrix[:, keyword_index].toarray().flatten()

# 각 문장에서의 TF-IDF 값 출력
for i, sentence in enumerate(sentences):
    if tfidf_values[i] > 0:
        print(f"Sentence {i + 1}: {sentence}, TF-IDF: {tfidf_values[i]}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# 예시 데이터
keywords = df[df['검색어'] == '빅데이터']
contents = df['contents'].dropna().tolist()
# TF-IDF 벡터화
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([*keywords, *contents])

# 키워드와 컨텐츠 간의 코사인 유사도 계산
cosine_similarities = cosine_similarity(tfidf_matrix)

# 결과 출력
print("Cosine Similarities:")
print(cosine_similarities)
```
