```python
import re

def remove_special_characters(text):
    # ì´ëª¨í‹°ì½˜ ë° íŠ¹ìˆ˜ë¬¸ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹
    pattern = r'[^a-zA-Z0-9ê°€-í£\s]'  # ì˜ë¬¸, í•œê¸€, ìˆ«ì, ê³µë°±ë§Œ í—ˆìš©

    # ì •ê·œí‘œí˜„ì‹ì„ ì‚¬ìš©í•˜ì—¬ íŠ¹ìˆ˜ë¬¸ì ì œê±°
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text

# 'ë‚´ìš©' ì—´ì˜ ê° í–‰ì— ëŒ€í•´ ì´ëª¨í‹°ì½˜ê³¼ íŠ¹ìˆ˜ë¬¸ì ì œê±°
df['contents'] = df['ë‚´ìš©'].dropna().apply(remove_special_characters)
```

# TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸
sentences = df['contents'].dropna().tolist()

# TF-IDF ë²¡í„°í™”
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(sentences)

# íŠ¹ì • í‚¤ì›Œë“œì˜ TF-IDF ê°’ì„ í™•ì¸
keyword_index = vectorizer.vocabulary_['ë¹…ë°ì´í„°']
tfidf_values = tfidf_matrix[:, keyword_index].toarray().flatten()

# ê° ë¬¸ì¥ì—ì„œì˜ TF-IDF ê°’ ì¶œë ¥
for i, sentence in enumerate(sentences):
    if tfidf_values[i] > 0:
        print(f"Sentence {i + 1}: {sentence}, TF-IDF: {tfidf_values[i]}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# ì˜ˆì‹œ ë°ì´í„°
keywords = df[df['ê²€ìƒ‰ì–´'] == 'ë¹…ë°ì´í„°']
contents = df['contents'].dropna().tolist()
# TF-IDF ë²¡í„°í™”
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform([*keywords, *contents])

# í‚¤ì›Œë“œì™€ ì»¨í…ì¸  ê°„ì˜ ì½”ì‚¬ì¸ ìœ ì‚¬ë„ ê³„ì‚°
cosine_similarities = cosine_similarity(tfidf_matrix)

# ê²°ê³¼ ì¶œë ¥
print("Cosine Similarities:")
print(cosine_similarities)
```
