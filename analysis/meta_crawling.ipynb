```python


# 데이터를 저장할 빈 리스트 생성
titles = []
contents = []
image_urls = []
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
month = 3
year = 3
max_failures = 3
failures = 0
num = 3

# 반복문 시작
for i in range(1, 50):  
    
    if i % 2 == 0:
        body_element = driver.find_element(By.TAG_NAME, 'body')
        body_element.send_keys(Keys.PAGE_DOWN)
    elif i % 30 == 0:
        driver.implicitly_wait(5)

    elif i == 50:
        month += 1
        i = 1  # i를 1로 리셋
   
    # 타이틀 추출

    try:
    
            print(f"{i}타이틀추출")
            title_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[1]/div/div/div/a/span'
            title_element = WebDriverWait(driver, 2).until(
                    EC.presence_of_element_located((By.XPATH, title_xpath))
                )
            title = title_element.text
            titles.append(title)
            driver.implicitly_wait(5)
            print(f"{i}타이틀추출완료")
        
    
    except TimeoutException:
        try: 
            title_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[1]/div/div/div/div/a/span'
            title_element = WebDriverWait(driver, 2).until(
                    EC.presence_of_element_located((By.XPATH, title_xpath))
                )
            title = title_element.text
            titles.append(title)
            driver.implicitly_wait(5)
            

        except TimeoutException:
            failures += 1
            print(f"{i}번 크롤링 실패 ({failures}/{max_failures})")
        
            # 3번 이상 실패하면 month를 증가시키고 실패 횟수 초기화
            if failures >= max_failures:
                month += 1
                num += 1
                print(f"month가 {month}로 증가되었습니다.")
                failures = 0
                i= 1
                
            continue
   
    

    # 본문 추출
    try:
        
        print(f"{i}본문 추출")
        content_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[2]/div/span/div/div/div'
        content_element = WebDriverWait(driver, 2).until(
            EC.presence_of_element_located((By.XPATH, content_xpath))
        )
        content = content_element.text
        contents.append(content)
        driver.implicitly_wait(2)
        print(f"{i}본문 추출완료")
        
    except TimeoutException:
        try :
             
            content_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[2]/div/span/div/div/div/span'
            content_element = WebDriverWait(driver, 2).until(
            EC.presence_of_element_located((By.XPATH, content_xpath))
            )
            content = content_element.text
            contents.append(content)
            driver.implicitly_wait(2)
            print(f"{i}본문 추출완료")
            
        except TimeoutException:
            print(f"{i}본문 없음 - 패스")
            
            continue

        continue
   
    # 이미지 파일 추출
    
    try:
        print(f"{i}이미지 파일 추출")
        img_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[2]/a/div[1]/img'
        img_element = WebDriverWait(driver, 2).until(
            EC.presence_of_element_located((By.XPATH, img_xpath))
        )
        img_url = img_element.get_attribute('src')
        image_urls.append(img_url)
        print(f"{i}이미지 파일 추출완료")
        # 이미지 다운로드
        print(f"{i}이미지 다운로드")
        response = requests.get(img_url)
        driver.implicitly_wait(2)
    except:
            driver.implicitly_wait(1)
            try :
                    
                video_xpath = f'//*[@id="content"]/div/div[1]/div/div[5]/div[2]/div[{month}]/div[{year}]/div[1]/div[{i}]/div/div[3]/div/div/div[2]/div[2]/div/div/div/div/div/div/video'
                video_element = WebDriverWait(driver, 2).until(
                    EC.presence_of_element_located((By.XPATH, video_xpath))
                    )
                video_url = video_element.get_attribute('src')
                video_urls.append(video_url)
                print(f"{i}비디오 파일 추출 완료")

                for i, video_url in enumerate(video_urls):
                    try:
                        print(f"{i}비디오 다운로드")
                        video_response = requests.get(video_url)
                        with open(f'video/video_{month}_{i}.mp4', 'wb') as video_file:
                            video_file.write(video_response.content)
                        print(f"{i}비디오 다운로드 완료")
                    except Exception as e:
                        print(f"{i}비디오 다운로드 실패: {str(e)}")
                        continue                
            except: 
                print(f"{i}비디오 파일 추출 실패")
                continue

            continue

        

    img = Image.open(BytesIO(response.content))
    img.save(f'images3/image_{num}{i}.jpg')
    print(f"{i}이미지 다운로드 완료")
   
               
driver.quit()

# DataFrame 생성
data = {'Title': titles, 'Content': contents}
df_3 = pd.DataFrame(data)

# DataFrame 출력 또는 다른 처리 수행
print(df_3)
```
