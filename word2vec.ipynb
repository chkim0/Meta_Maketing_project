```python
import re

def remove_special_characters(text):
    # ì´ëª¨í‹°ì½˜ ë° íŠ¹ìˆ˜ë¬¸ì ì œê±°ë¥¼ ìœ„í•œ ì •ê·œí‘œí˜„ì‹
    pattern = r'[^a-zA-Z0-9ê°€-í£\s]'  # ì˜ë¬¸, í•œê¸€, ìˆ«ì, ê³µë°±ë§Œ í—ˆìš©

    # ì •ê·œí‘œí˜„ì‹ì„ ì‚¬ìš©í•˜ì—¬ íŠ¹ìˆ˜ë¬¸ì ì œê±°
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text

# 'ë‚´ìš©' ì—´ì˜ ê° í–‰ì— ëŒ€í•´ ì´ëª¨í‹°ì½˜ê³¼ íŠ¹ìˆ˜ë¬¸ì ì œê±°
df['contents_sentense'] = df['contents'].dropna().apply(remove_special_characters)
```

## ë¬¸ì¥ì—ì„œ ë‹¨ì–´ ìœ ì‚¬ë„ Word2Vec

```python
# ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸
sentences = df['contents'].dropna().tolist()

# ë¬¸ì¥ì„ ë‹¨ì–´ë¡œ í† í°í™”
tokenized_sentences = [word_tokenize(sentence.lower()) for sentence in sentences]

# Word2Vec ëª¨ë¸ í›ˆë ¨
model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)

# íŠ¹ì • í‚¤ì›Œë“œì™€ ìœ ì‚¬í•œ ë‹¨ì–´ ì°¾ê¸°
similar_words = [(word, "{:.4f}".format(score)) for word, score in model.wv.most_similar('ë¶€íŠ¸ìº í”„', topn=10) if len(word) > 1]

# ìœ ì‚¬í•œ ë¬¸ì¥ ì°¾ê¸°
unique_sentences = list(set(sentences))
similar_sentences = [sentence for sentence in unique_sentences if 'ë¶€íŠ¸ìº í”„' in sentence.lower()]

# ê²°ê³¼ ì¶œë ¥
print(f"Similar words: {similar_words}")
print(f"Similar sentences: {similar_sentences}")
```
